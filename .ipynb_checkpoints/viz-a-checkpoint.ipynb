# Data Visualization

## Data Scientist Job Market in the U.S.

```python
import seaborn as sns
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

### loading data from the CSV file for the Data Scientist Job Market in the U.S.

```python
data = pd.read_csv("alldata.csv")
data
```

### View the top 20 company and their location by their position

```python
sns.set_theme()
data1=data.head(20)
sns.displot(x="company",y="location",data=data1,hue="position")
```

### Find highest 10 position in location 

```python
location_data=data.set_index("position").sort_index(ascending=False).head(10)
location_data
sns.relplot(x="location",y="position",data=location_data,hue="company")
```
