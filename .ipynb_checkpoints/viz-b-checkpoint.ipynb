# Data Visualization

## Top 250 Football transfers from 2000 to 2018

```python
import seaborn as sns
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

### loading data from the CSV file for the Top 250 Football transfers from 2000 to 2018¶

```python
football_data = pd.read_csv("top250-00-19.csv")
football_data
```

```python
sns.set_theme()
sns.pairplot(data=football_data, hue="Season")
```

### Summary

 1. Most of the players are between the ages of 20-30 years old
 2. As the years went by, Transfer_fee decreased and Market_value increased
